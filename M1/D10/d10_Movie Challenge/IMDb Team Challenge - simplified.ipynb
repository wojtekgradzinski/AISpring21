# IMDB Team Challenge

It consists of scraping the following information from IMDB:

* Movie name
* Description
* Release Date
* Director Name
* Rating
* Duration
* Genre
* Stars (Actors)
* Filming Dates

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup


# request url
page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=adventure&sort=user_rating,desc&start=1")

# create a BeautifulSoup object
soup = BeautifulSoup(page.content, 'html.parser')

# page title
title = soup.find('title')
display(title.string)
# data['title'] = title.string

# - Titles
titles_container = soup.find_all(class_= "lister-item-header")
# display(titles_container)
titles = [title.get_text() for title in titles_container ] #loop to getText in days_container
# display(titles)
titles = [title.split('\n') for title in titles] #split by \n
# display(titles)
titles_names = [title[2] for title in titles] #split by \n
# display(titles_names)

# - Chart number
chart_number = [title[1] for title in titles] #split by \n
chart_number = [title.replace('.','') for title in chart_number] #split by \n
# display(chart_number)

# - Released year
published_year = [title[3] for title in titles]
# display(titles)

# - Director
movies_director=soup.find_all(class_= "lister-item-content")
director_list=[]
for director in movies_director:
    director_list.append(director.find("p",{"class":""}).a.text)
    
# - Duration
duration = soup.find_all(class_='lister-item-content')
dur = [title.get_text() for title in duration]
dur = [title.split('\n') for title in dur]
#display(dur)
duration = [title[9] for title in dur]
# display(duration)

# - Genre
genre_container = soup.find_all(class_='genre')
gr = [title.get_text() for title in genre_container]
gr = [title.split('\n') for title in gr]
genre = [title[1] for title in gr]


# - Rating
movies_rating=soup.find_all(class_="inline-block ratings-imdb-rating")
rating_list=[]
for rating in movies_rating:
    rating_list.append(rating.strong.text)



df = pd.DataFrame({'Chart Number': chart_number,
                    'Movie Titles': titles_names, 
                    'Released Date': published_year,
                    'Director':director_list,
                    'Stars': stars,
                    'Duration':duration,
                    'Genre':genre,
                    'Rating':rating_list})
df


# page_details = requests.get("https://www.imdb.com/title/tt0167260/locations?ref_=tt_dt_dt")
# soup = BeautifulSoup(page_details.content, 'html.parser')

# filming_dates = soup.find_all("li",{'id': "filming_dates", 'class': "ipl-zebra-list"})
# filming_dates_list = [i.text.split(",")]
# # filming_dates_list = [film.split("\n") for film in filming_dates_list] #split by \n
# display(filming_dates_list)


# # df = pd.DataFrame({'Filming Dates': filming_dates,
# #                     })
# # df

```
